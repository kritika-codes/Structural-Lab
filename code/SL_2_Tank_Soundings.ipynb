```python
import numpy as np
import pandas as pd
```

```python
inner_df=pd.read_excel('inner.xlsx')
outer_df=pd.read_excel('outer.xlsx')
```

```python
z_values = inner_df.iloc[:, 0].to_numpy()
print(z_values)
```

```python
def calculate_Awp(df):
 Awp=np.zeros(len(df))

 for i, (z_value, row) in enumerate(df.iterrows()):
    x_values = df.columns.values[1:].astype(float)
    y_values = row.values[1:]
    Awp[i] = np.trapz(y_values, x_values)
    print(f"Awp at z={row.values[0]}: {Awp[i]}")

 Awp[0] = 0 if np.isnan(Awp[0]) else Awp[0]
 Awp[-1] = 0 if np.isnan(Awp[-1]) else Awp[-1]
 return Awp
```

```python
Awp_inner=calculate_Awp(inner_df)
Awp_outer=calculate_Awp(outer_df)
Awp=Awp_outer-Awp_inner
```

```python
def calculate_sectional_areas(df):
  x_locations = df.columns[1:].astype(float)
  sectional_area_table = []
  z_values = df.iloc[:, 0].to_numpy()
  for x in x_locations:
    y_values = df[x].to_numpy()
    cumulative_area = np.zeros_like(z_values)


    for i in range(len(z_values)):
        cumulative_area[i] = np.trapz(y_values[:i+1], z_values[:i+1])

    sectional_area_table.append(cumulative_area)

  sectional_area_df = pd.DataFrame(
    data=np.array(sectional_area_table).T,
    columns=x_locations,
    index=z_values
  )
  return sectional_area_df
```

```python
inner_section_df=calculate_sectional_areas(inner_df)
outer_section_df=calculate_sectional_areas(outer_df)
sectional_area_df=outer_section_df.sub(inner_section_df)
inner_section_df.to_excel('inner_section.xlsx')
outer_section_df.to_excel('outer_section.xlsx')
```

```python
def calculate_volume(df,Awp):
 vol=np.zeros(len(df))
 z_values = df.iloc[:, 0].to_numpy()
 for i in range(len(df)):
   vol[i]=np.trapz(Awp[:i+1],z_values[:i+1])
   vol[i]=vol[i].round(4)
   print(f"Displacement at z={z_values[i]}: {vol[i]}")

 vol[0] = 0 if np.isnan(vol[0]) else vol[0]
 vol[-1] = 0 if np.isnan(vol[-1]) else vol[-1]
 return vol
```

```python
inner_volume=calculate_volume(inner_df,Awp_inner)
outer_volume=calculate_volume(outer_df,Awp_outer)
vol=outer_volume-inner_volume
```

```python
def calculate_load_percentage(vol):
 load=np.zeros(len(vol))
 load=(vol/vol.max())*100
 load=load.round(4)
 for i in range(len(vol)):
   print(f"Load percentage at z={z_values[i]}: {load[i]}")

 return load
```

```python
load=calculate_load_percentage(vol)
```

```python
def calculate_yc(df):
  centroids=[]
  x_locations = df.columns[1:].astype(float)
  z_values = df.iloc[:, 0].to_numpy()

  for x in x_locations:
    y_values = df[x].to_numpy()
    centroid = np.zeros_like(z_values)

    for i in range(len(z_values)):
        centroid[i] = np.trapz((y_values[:i+1]**2)/2, z_values[:i+1])/np.trapz(y_values[:i+1], z_values[:i+1])
        if np.isnan(centroid[i]):
               centroid[i] = 0

    centroids.append(centroid)

  centroids_df = pd.DataFrame(
    data=np.array(centroids).T,
    columns=x_locations,
    index=z_values
  )
  return centroids_df
```

```python
inner_centroids_df=calculate_yc(inner_df)
outer_centroids_df=calculate_yc(outer_df)

df1=inner_centroids_df.mul(inner_section_df)
df2=outer_centroids_df.mul(outer_section_df)
df3=df2.sub(df1)
centroids_df=df3.div(sectional_area_df)
if centroids_df.isna().any().any():
  centroids_df=centroids_df.fillna(0)
print(centroids_df.to_string())
```

```python
def calculate_tcg(df,centroids_df,sectional_area_df, vol):
  tcg=np.zeros(len(df))

  for i, (z_value, row) in enumerate(df.iterrows()):
    x_values = df.columns.values[1:].astype(float)
    sectional_area=sectional_area_df.iloc[i].to_numpy()
    centroids=centroids_df.iloc[i].to_numpy()
    tcg[i] = np.trapz(centroids * sectional_area, x_values)/vol[i]
    if np.isnan(tcg[i]):
      tcg[i]=0
    print(f"tcg at z={row.values[0]}: {tcg[i]}")

  return tcg
```

```python
tcg_inner=calculate_tcg(inner_df,inner_centroids_df,inner_section_df,inner_volume)
tcg_outer=calculate_tcg(outer_df,outer_centroids_df,outer_section_df,outer_volume)

tcg=(tcg_outer*outer_volume-tcg_inner*inner_volume)/vol
tcg = np.nan_to_num(tcg, nan=0.0)
```

```python
def calculate_vcg(df, Awp, vol):
 vcg=np.zeros(len(df))
 z_values = df.iloc[:, 0].to_numpy()
 arr=Awp*z_values
 for i in range(len(df)):
   vcg[i]=np.trapz(arr[:i+1],z_values[:i+1])/vol[i]
   if np.isnan(vcg[i]):
     vcg[i]=0
   print(f"vcg at z={z_values[i]}: {vcg[i]}")

 vcg[0] = 0 if np.isnan(vcg[0]) else vcg[0]
 return vcg
```

```python
vcg_inner=calculate_vcg(inner_df,Awp_inner,inner_volume)
vcg_outer=calculate_vcg(outer_df,Awp_outer,outer_volume)

vcg=(vcg_outer*outer_volume-vcg_inner*inner_volume)/vol
vcg = np.nan_to_num(vcg, nan=0.0)
```

```python
def calculate_lcg(df,sectional_area_df,vol):
 lcg=np.zeros(len(df))

 for i, (z_value, row) in enumerate(df.iterrows()):
    x_values = df.columns.values[1:].astype(float)
    sectional_area=sectional_area_df.iloc[i].to_numpy()
    lcg[i] = np.trapz(x_values * sectional_area, x_values)/vol[i]
    if np.isnan(lcg[i]):
      lcg[i]=0
    print(f"lcg at z={row.values[0]}: {lcg[i]}")

 lcg[0]=0 if np.isnan(lcg[0]) else lcg[0]
 return lcg
```

```python
lcg_inner=calculate_lcg(inner_df,inner_section_df,inner_volume)
lcg_outer=calculate_lcg(outer_df,outer_section_df,outer_volume)

lcg=(lcg_outer*outer_volume-lcg_inner*inner_volume)/vol
lcg = np.nan_to_num(lcg, nan=0.0)
```

```python
soundings=pd.DataFrame({
    'Depth (m)':z_values,
    'Load(%)':load,
    'Volume (m^3)':vol,
    'LCG (m)':lcg,
    'VCG (m)':vcg,
    'TCG (m)':tcg
})

print(soundings.to_string(index=False))
```

```python
soundings.to_excel('soundings.xlsx')
```
